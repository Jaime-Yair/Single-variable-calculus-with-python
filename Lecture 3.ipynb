```python
x = Symbol('x')
expres = sin(x)/x
expres
```

```python
limit(expres, x, 0)
```

```python
expres = (cos(x)-1)/x
expres
```

# Plotting the derivatives

```python
def f(x):
    return np.sin(x)

def deriv(x):
    return derivative(f,x)


x=np.linspace(-5,5,100)

plt.plot(x,np.sin(x), label= 'function')
plt.plot(x, deriv(x), label= 'derivative')

plt.legend()
```

```python
def sin(x):
    return np.cos(x)

def deriv(x):
    return derivative(f,x)


x=np.linspace(-5,5,100)

plt.plot(x,sin(x), label= 'function')
plt.plot(x,deriv(x), label= 'derivative')

plt.legend()
```
