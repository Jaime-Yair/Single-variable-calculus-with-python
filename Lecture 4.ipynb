```python
from sympy import *
import matplotlib.pyplot as plt
from scipy.misc import derivative
import numpy as np
from sympy import limit
```

#### Quotient formula

```python
def f(x):
    return np.sin(x)/np.cos(x)

def deriva(x):
    return derivative(f,x)

x = np.linspace(-10-10,100)

plt.plot(x,f(x), label = 'f(x)')
plt.plot(x, deriva(x), label= 'fÂ´(x)')
plt.legend()
```

```python
x = Symbol('x')
expression = sin(x)/cos(x)
diff(expression)
```

#### Product formula

```python
def f(x):
    return np.sin(x)*np.cos(x)

def deriva(x):
    return derivative(f,x)

x = np.linspace(-10-10,100)

plt.plot(x,f(x), label = 'f(x)')
plt.plot(x, deriva(x), label= 'fÂ´(x)')
plt.legend()
```

```python
x = Symbol('x')
expression = sin(x)*cos(x)
diff(expression)
```

###### Chain Rule

```python
x, t = symbols('x t')
fx = sin(x)
gt = t**2 
```

```python
diff(fx), diff(gt)
```

```python
fx.diff(x), gt.diff(t)
```

```python
diff(sin(t**2))
```
