```python
import sympy as sp
from sympy import Symbol, diff
import matplotlib.pyplot as plt
from scipy.misc import derivative
import numpy as np
```

```python
def f(x):
    return 1/(1+x**2)

x = np.linspace(-3,3,100)

plt.plot(x,f(x))
```

```python
x = Symbol('x')
diff(f(x))
```

```python
# the derivative

def f(x):
    return 1/(1+x**2)

def deriv(x):
    return derivative(f, x)
    

x = np.linspace(-15,15,100)

plt.plot(x,f(x),label='function')
plt.plot(x,deriv(x), label='derivative')
plt.axvline(x=0, alpha=0.2, linestyle='--')
plt.axhline(y=0, alpha=0.2, linestyle='--')
```

```python
# Define function
def f(x): 
    return 1/(1+x**2)

# Define the derivative
def slope(x): 
    return derivative(f, x)

# Define x data range 
x = np.linspace(-5,5,50)

# Choose point to plot tangent line
x1 = 1
y1 = f(x1)

m  = slope(x1)

print("cordinates of the tangent: ",(x1,y1)) 
print("The slope at the chosen point: ", m)

# Define tangent line
# y = m*(x - x1) + y1
def line(x, x1, y1):
    return m*(x - x1) + y1

# Define x data range for tangent line
xrange = np.linspace(x1-1, x1+1, 10)

# Plot the figure
plt.figure()
plt.plot(x, f(x))
plt.scatter(x1, y1, color='C1', s=50)
plt.plot(xrange, line(xrange, x1, y1), 'C1--', linewidth = 2)
plt.axvline(x=0, alpha=0.2, linestyle='--')
plt.axhline(y=0, alpha=0.2, linestyle='--')
```
