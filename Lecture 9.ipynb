```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.misc import derivative
from sympy import *
```

```python
def f(x):
    return np.log(x)

x = np.linspace(0.1,5,100)

plt.plot(x,f(x))
```

```python
var('x')
diff(ln(x))
```

```python
def deri(x):
    return 1/x
```

```python
x_0 = 1
print(f(x_0))
print(deri(x_0))
```

```python
def f(x):
    return np.log(x)

def g(x):
    return x-1
    
x = np.linspace(0.1,5,100)

plt.plot(x,f(x))
plt.plot(x,g(x))
```

# Using polifit

```python
x = np.arange(0.1, 10, 0.01)
y = np.log(x)

p1 = np.polyfit(x,y, 1)
p2 = np.polyfit(x,y, 3)

y1 = np.polyval(p1, x)
y2 = np.polyval(p2, x)


plt.plot(x, y, label='Log(x)')
plt.plot(x, y1, label='Degree 1')
plt.plot(x, y2, label='Degree 2')
plt.legend()

plt.show()
```

### Quadratic approximation

```python
def f(x):
    return np.cos(x)

def deriva(x):
    return 1-(x**2)/2


x = np.linspace(-5,5,1000)


plt.plot(x,f(x), label='f(x)')
plt.plot(x,deriva(x), label='2 approximation')
plt.legend()
```
