```python
from sympy import Symbol, diff, sin, cos, tan
import matplotlib.pyplot as plt
from scipy.misc import derivative
import numpy as np
from sympy import limit
```

# Example

```python
def f(t):
    return 80-5*t**2 


x = np.linspace(-3,3,100)

plt.plot(x,f(x))
```

```python
t = Symbol('t')
diff(f(t))
```

```python
# the derivative

def f(t):
    return 80-5*t**2 

def deriv(t):
    return derivative(f, t)
    

t = np.linspace(-5,5,100)

plt.plot(t,f(t), label="function")
plt.plot(t,deriv(t),label="derivate")
plt.axvline(x=0, alpha=0.2, linestyle='--')
plt.axhline(y=0, alpha=0.2, linestyle='--')

plt.legend()
plt.show()
```

```python
# Define function
def f(t): 
    return 80-5*t**2 

# Define the derivative
def slope(t): 
    return derivative(f, t)

# Define x data range 
t = np.linspace(-5,5,50)

# Choose point to plot tangent line
t1 = 3
y1 = f(t1)

m  = slope(t1)

print("cordinates of the tangent: ",(t1,y1)) 
print("The slope at the chosen point: ", m)

# Define tangent line
# y = m*(x - x1) + y1
def line(t, t1, y1):
    return m*(t - t1) + y1

# Define x data range for tangent line
xrange = np.linspace(t1-1, t1+1, 10)

# Plot the figure
plt.figure()
plt.plot(t, f(t))
plt.scatter(t1, y1, color='C1', s=50)
plt.plot(xrange, line(xrange, t1, y1), 'C1--', linewidth = 2)
plt.axvline(x=0, alpha=0.2, linestyle='--')
plt.axhline(y=0, alpha=0.2, linestyle='--')
```

# Example piecewise function

```python
def f(x):
    return np.piecewise(x, [x >= 0, x < 0], [lambda x : x+1, lambda x : -x + 2])

x = np.linspace(-2,2,101)

plt.plot(x,f(x))
plt.axvline(x=0, alpha=0.2, linestyle='--')
plt.axhline(y=0, alpha=0.2, linestyle='--')
plt.scatter(0, 1, color='C1', s=40)
plt.scatter(0, 2, color='C1', s=40)
```

# Two Trig Limits

#### First example

```python
def f(x):
    return np.sin(x)/x

def derivate(x):
    return derivative(f,x)

x = np.linspace(-15,15,1000)

plt.plot(x,f(x),label="function")
plt.plot(x,derivate(x),label="derivate")
plt.legend()
plt.show()
```

```python
x = Symbol('x')
```

```python
expression= sin(x)/x
```

```python
diff(expression)
```

```python
limit(expression,x,0)
```

#### Second example

```python
def f(x):
    return (1-np.cos(x))/x

def derivate(x):
    return derivative(f,x)

x = np.linspace(-15,15,1000)

plt.plot(x,f(x),label="function")
plt.plot(x,derivate(x),label="derivate")

plt.legend()
plt.show()
```

```python
x = Symbol('x')

expression = (1-cos(x))/x
```

```python
diff(expression)
```

```python
limit(expression,x, 0)
```
