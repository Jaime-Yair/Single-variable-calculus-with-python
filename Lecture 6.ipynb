```python
from sympy import *
import matplotlib.pyplot as plt
from scipy.misc import derivative
import numpy as np
```

```python
# Define function
def f(x): 
    return 2**x

# Define the derivative
def slope(x): 
    return derivative(f, x)

# Define x data range 
x = np.linspace(-2,2,50)

# Choose point to plot tangent line
x1 = 0
y1 = f(x1)

m  = slope(x1)

print("cordinates of the tangent: ",(x1,y1)) 
print("The slope at the chosen point: ", m)

# Define tangent line
# y = m*(x - x1) + y1
def line(x, x1, y1):
    return m*(x - x1) + y1

# Define x data range for tangent line
xrange = np.linspace(x1-1, x1+1, 10)

# Plot the figure
plt.figure()
plt.plot(x, f(x))
plt.scatter(x1, y1, color='C1', s=50)
plt.plot(xrange, line(xrange, x1, y1), 'C1--', linewidth = 2)
plt.axvline(x=0, alpha=0.2, linestyle='--')
plt.axhline(y=0, alpha=0.2, linestyle='--')
```

```python
# Define function
def f(t): 
    return np.power(2, t)

# Define the derivative
def slope(t): 
    return derivative(f, t)

# Define x data range 
t = np.linspace(-5,5,50)

# Choose point to plot tangent line
t1 = [0,2,3]  #Change any of these points to see the graph
y1 = f(t1)

m  = [slope(i) for i in t1]    

for i in range(len(t1)):
    print("cordinates of the tangent for point",i+1,':   ',(t1[i],y1[i]))
    
for i in range(len(t1)):
    print("The slope at the chosen point",i+1,':   ', m[i].round(4))



# Define x data range for tangent line
xranges=[]
for i in range(len(t1)):
    xranges.append(np.linspace(t1[i]-1, t1[i]+1, 10)) 


# Define tangent line
# y = m*(x - x1) + y1
lines = []
for i in range(len(t1)):
    lines.append(m[i]*(xranges[i] - t1[i]) + y1[i])


# Plot the figure
plt.figure()
plt.plot(t, f(t))  # Plot the function
for i in range(len(t1)):
    plt.scatter(t1[i], y1[i], color='C1', s=50)   # Plot the points where the slope is 0
for i in range(len(t1)):
    plt.plot(xranges[i], lines[i], 'C1--', linewidth = 2)   # Plot the lines
plt.axvline(x=0, alpha=0.2, linestyle='--')
plt.axhline(y=0, alpha=0.2, linestyle='--')
```
