```python
from sympy import *
import matplotlib.pyplot as plt
from scipy.misc import derivative
import numpy as np
```

##### Implicit differentiation


```python
#Example 2

var('x y')
formula = x**2 + y**2 - 1   # this represents x^2+y^2=1
idiff( formula, y, x )
```

```python
def f(x):
    return np.sqrt(1-x**2)

def deriv(x):
    return derivative(f,x)

x = np.linspace(-1,1,1000)

plt.plot(x, f(x))
```

```python
def f(x):
    return np.sqrt(x)

def deriv(x):
    return derivative(f,x)

x = np.linspace(1,10,1000)

plt.plot(x, f(x))
plt.plot(x, deriv(x))
```

```python
def tan(x):
    return np.tan(x)

def tan_inverse(x):
    return np.arctan(x)

x = np.linspace(-2 * np.pi, 2 * np.pi, 1000)

plt.plot(x,tan(x), label='tan(x)')
plt.plot(x,tan_inverse(x), label='arc tan(x)')
plt.legend()
plt.ylim(-5,5)

a = np.pi/2

plt.axvline(x=0, alpha=0.2, linestyle='--')
plt.axhline(y=a, alpha=0.2, linestyle='--')
plt.axhline(y=-a, alpha=0.2, linestyle='--')
plt.axhline(y=0, alpha=0.2, linestyle='--')
```
